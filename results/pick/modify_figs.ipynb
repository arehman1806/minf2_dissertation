```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the data from a CSV file
data_path = './old_rotational_slip_constant_data.csv'  # Replace with your actual file path
df = pd.read_csv(data_path)
df.loc[df['Step'] >= 290, ['Delta Roll', 'Delta Pitch', 'Delta Yaw']] = 0

# Plotting
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(5, 10))

# First plot for Delta Roll, Delta Pitch, and Delta Yaw
ax1.plot(df['Step'], df['Delta Roll'], label='Δ Roll')
ax1.plot(df['Step'], df['Delta Pitch'], label='Δ Pitch')
ax1.plot(df['Step'], df['Delta Yaw'], label='Δ Yaw')
ax1.set_ylabel('first order derivative of orientation (rad/unit sim time)')
ax1.legend()
ax1.set_ylim(-0.05, 0.05)
ax1.set_xlim(0, 300)
ax1.axvline(x=290, color='r', linestyle='--')
ax1.text(290, 0.03, 'Lost Contact', rotation=90)


# Second plot for Force
ax2.plot(df['Step'], df['Force'], label='Force', color='magenta')
ax2.set_xlabel('Step')
ax2.set_ylabel('Force (N)')
ax2.set_ylim(-10, 10)
ax2.set_xlim(0, 300)
ax2.axvline(x=290, color='r', linestyle='--')
ax2.text(290, 5, 'Lost Contact', rotation=90)
ax2.legend()

# Adjust layout to prevent overlap and ensure labels are displayed correctly
plt.tight_layout()

# Save the modified plot
plt.savefig('modified_rotational_slip_constant_plot.png')

# Show the plot
plt.show()
```
